```python
df_new = pd.get_dummies(df, columns=["Kasa Tipi","Vites Türü","Yakıt Türü"], dtype=int, drop_first=True)
```

```python
y = df_new[["Fiyat"]]
x = df_new.drop("Fiyat", axis  =1)
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
model.predict([[2017,30000,1.6,110,1,2600,0,1]])
```

```python
model.score(x,y)
```
