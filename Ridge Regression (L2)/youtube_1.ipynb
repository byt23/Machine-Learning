```python
# y = (a1 * x1) + (a2 * x2) + ... + (an * xn) + b + alfa *(katsayılar toplamı) ** 2

# 50 = 40 + 10 + 0
# 50 = 30 + 10 + 10
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression, Ridge
```

```python
df = pd.read_csv("student_scores.csv")
df.head(3)
```

```python
y = df["Scores"]
x = df[["Hours"]]
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(6,6))
plt.scatter(x,y)
plt.show()
```

```python
lr = LinearRegression()
model = lr.fit(x,y)
model.score(x,y)
```

```python
alfalar = [1,10,20,100,200]
for a in alfalar :
    r = Ridge(alpha=a)
    model_r = r.fit(x,y)
    skor = model_r.score(x,y)
    print("Skor :",skor)
    print("Katsayı :", model.coef_)
```
