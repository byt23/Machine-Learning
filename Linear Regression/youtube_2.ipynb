```python
from sklearn.linear_model import LinearRegression
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("Student_Marks.csv")
```

```python
df.head(3)
```

```python
df.columns = ["sınıf", "saat", "puan"]
```

```python
df.head(3)
```

```python
y = df[["puan"]]
x = df[["sınıf","saat"]]
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
# y = ax + b
```

```python
model.coef_ # katsayı
```

```python
model.intercept_ # sabit 
```

```python
model.predict([[4,5]])
```

```python
(4 * 1.86405074) + (5 * 5.39917879) -7.45634623 # proof of predict calculation
```

```python
df.head(3)
```

```python
s1 = model.predict([[3,4.508]])
s1
```

```python
abs(19.202 - s1) / 19.202  # %17'lik bir sapma olduğunu gördük, ve abs fonksiyonunu koyarak mutlak değerini aldım.
```

```python
model.score(x,y) # %6'lık sapma genel anlamda olduğunu gördük.
```
