```python
import pandas as pd  
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Audi.csv")
```

```python
df.head(3)
```

```python
df = df[["Year","Type","Mileage(miles)","Engine","PS","Transmission","Fuel","Number_of_Owners","Price(£)"]]
```

```python
df.head(3)
```

```python
df.columns = ["Yıl","Araba Türü","Mil","Motor Hacmi","Beygir Gücü","Vites Türü","Yakıt Türü","Sahip Sayısı","Fiyat"]
```

```python
df.head(3)
```

```python
df["Motor Hacmi"] = df["Motor Hacmi"].str.replace("L","")
```

```python
df["Motor Hacmi"] = df["Motor Hacmi"].astype(float)
```

```python
df = pd.get_dummies(df, columns = ["Araba Türü", "Vites Türü", "Yakıt Türü"],dtype=int, drop_first=True)
```

```python
df.head(3)
```

```python
df["Yıl"] = df["Yıl"].astype("int")
```

```python
df.head()
```

```python
y = df[["Fiyat"]]
x = df.drop("Fiyat", axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.80,random_state=22)
```

```python
lm = LinearRegression()
model = lm.fit(x_train,y_train)
model.score(x_test,y_test)
```
